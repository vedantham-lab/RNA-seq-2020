```r
new_VVCM = VVCM[VVCM$ENSEMBLID %in% devCM$Gene_ENSEMBL,]
merge = length(new_VVCM$ENSEMBLID)
png(file = "venn_CM.png")
venn.plot = draw.pairwise.venn((length(devCM$Gene_ENSEMBL)),(length(VVCM$ENSEMBLID)),merge,category = c("Dev2019","VV"), scaled=TRUE,  col = rep("black",
+ 2), fill = c("BLUE","RED"), cex = c(1.5,1.5,1.5), cat.pos = c(-45, 45), cat.dist = 0.033, cat.cex =1.5)
grid.draw(venn.plot)
dev.off()
```
